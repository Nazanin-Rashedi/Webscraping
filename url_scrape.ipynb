```python
url_df = pd.read_csv("------------------------")
```

```python
url_df.insert(loc=2, column='abstract', value ='')
url_df.insert(loc=3, column='headline', value ='')
url_df.insert(loc=4, column='text', value ='')
url_df.insert(loc=5, column='homepage', value ='')
```

```python
def find_homepage(row):
  return 1 if re.compile(r'www.|^'+row['domain']+'/?$').match(row['url']) else 0
```

```python
url_df['homepage'] = url_df.apply(find_homepage, axis=1)
```

```python
url_nothome_df = url_df[url_df['homepage']==0]

# results in 295 urls
```

```python
def get_achtgut(url):

  headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36'}
  req = requests.get("https://"+url, headers=headers)
  soup = BeautifulSoup(req.content, 'html.parser')
  text = ""
  head = ""
  try:
    head = soup.find('h2').text
    content = soup.find("div",{"class":"artikeltext"}) if soup.find("div",{"class":"artikeltext"})!=None else soup.find("div",{"class":"teaser_blog_text"})
  
    for p in content.children:
      if str(p).startswith('<p>'):
        text = text + str(p) + " "
    text = text.replace('<p>', '')
    text = text.replace('</p>', '')
    text = text.replace('<h3>', '')
    text = text.replace('</h3>', '')
    text = text.replace('<strong>', '')
    text = text.replace('</strong>', '')
    text = text.replace(u'\xa0', ' ')
    text = re.sub('<a.*>', '', text) 
  except:
    print("problem in "+url)

   

  return head,text
```

```python
for index, row in url_nothome_df[url_nothome_df['domain']=='achgut.com'].iterrows():
  a,b = get_achtgut(row['url'])
  url_nothome_df.at[index,'headline'] = a
  url_nothome_df.at[index,'text'] = b
```

```python
def get_sciencefiles(url):

  headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36'}
  req = requests.get("https://"+url, headers=headers)
  soup = BeautifulSoup(req.content, 'html.parser')
  text = ""
  head = ""
  try:
    head = soup.find('h1').text
    content = soup.find("div",{"class":"post-single-content box mark-links"}) 
    #if soup.find("div",{"class":"artikeltext"})!=None else soup.find("div",{"class":"teaser_blog_text"})
  
    for p in content.children:
      if str(p).startswith('<p>'):
        if p.find('span'):
          text = text + p.find('span').getText()
        else :
          text = text + str(p) + " "
    text = text.replace('<p>', '')
    text = text.replace('</p>', '')
    text = text.replace('<ul>', '')
    text = text.replace('</ul>', '')
    text = text.replace('<li>', '')
    text = text.replace('</li>', '')
    #text = text.replace(u'\xa0', ' ')
    text = re.sub('<a.*>', '', text) 
    text = re.sub('<img.*>', '', text) 
    text = re.sub('<input.*>', '', text) 
    text = re.sub('<form.*>', '', text) 
  except:
    print("problem in "+url)

   

  return head,text
```

```python
for index, row in url_nothome_df[url_nothome_df['domain']=='sciencefiles.org'].iterrows():
  a,b = get_sciencefiles(row['url'])
  url_nothome_df.at[index,'headline'] = a
  url_nothome_df.at[index,'text'] = b
```

```python
def get_deutsche(url):

  headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36'}
  req = requests.get("https://"+url, headers=headers)
  soup = BeautifulSoup(req.content, 'html.parser')
  text = ""
  head = ""
  abstract = ""
  if url.startswith("express.deutsche-wirtschafts-nachrichten.de"):
    try:
      outer = soup.find("div",{"class":"finlive-list-one wall tip_100"})
      head = outer.find('a').text
      abstract = soup.find("div",{"class":"flash"}).text
    except:
      print("express problem in "+url)
  try:
    head = soup.find('h1').text
    ps = soup.find_all("p",{"class":"artclass"}) 
    abstract = soup.find("div",{"class":"article-flash"}).text
    #if soup.find("div",{"class":"artikeltext"})!=None else soup.find("div",{"class":"teaser_blog_text"})
  
    for p in ps:
      if p.find('a')!=None:
        text = text + p.find('a').getText()
        text = text + str(p) + " "
    text = text.replace('<p>', '')
    text = text.replace('</p>', '')
    text = text.replace('<p class="artclass">', '')
    text = text.replace('<ul>', '')
    text = text.replace('</ul>', '')
    text = text.replace('<li>', '')
    text = text.replace('</li>', '')
    #text = text.replace(u'\xa0', ' ')
    text = re.sub('<a.*>', '', text) 
    text = re.sub('<img.*>', '', text) 
    text = re.sub('<input.*>', '', text) 
    text = re.sub('<form.*>', '', text) 
  except:
    print("problem in "+url)
  return head,text,abstract
```
